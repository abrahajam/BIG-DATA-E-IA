```python
# 1. INSTALACIÓN (Solo necesario en Colab)
# Ejecuta esta línea primero para instalar PySpark en la nube de Google
!pip install pyspark

# ---------------------------------------------------------

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum

# 2. CREAR DATOS DE EJEMPLO (Simulamos el archivo CSV)
# Como no tienes el archivo 'ventas.csv' en la nube, lo creamos al vuelo.
datos = [
    ("España", 1000),
    ("España", 2000),
    ("México", 1500),
    ("México", 500),
    ("Colombia", 1200)
]
columnas = ["Pais", "Precio"]

# 3. INICIAR SESIÓN (Arrancar el Clúster)
spark = SparkSession.builder.appName("ClaseBigData").getOrCreate()

# 4. CREAR DATAFRAME (El equivalente a leer el Excel)
df_ventas = spark.createDataFrame(datos, columnas)

# 5. PROCESAMIENTO (La Magia)
print("--- Procesando datos distribuidos ---")
resultado = df_ventas.groupBy("Pais") \
                     .agg(sum("Precio").alias("Total_Ventas"))

# 6. MOSTRAR RESULTADO
resultado.show()
```
