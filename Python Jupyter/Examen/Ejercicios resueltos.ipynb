### 2. Palíndromo

```python
palindromo="HooH"
palindromo2=[]
palindromo_espejo=[]

for i in palindromo:
    if(i != " "):
        palindromo2.append(i.lower())
        palindromo_espejo.insert(0,i.lower())


if palindromo2 == palindromo_espejo:
    print("Es un palíndormo")
    
else:
    print("No es un palindromo")
       
```

### 4. Conteo de Frecuencia de Palabras


```python
texto = ("w ww w w w w w w w")  

palabra_contador = {}

for palabra in texto.split():
    if palabra in palabra_contador:
        palabra_contador[palabra] +=1
    else:
        palabra_contador[palabra] = 1

print(f'{"Palabra":<12} Repeticiones')

for palabra,veces in sorted(palabra_contador.items()):
    print(f'{palabra:<12}{veces}')

print("Número de palabras únicas en el texto",len(palabra_contador))
```

### 5. Factorial con Recursión

```python
def fac_recur(n):
    if n == 0:
        return 1
    else:
        return n*fac_recur(n-1)

while True:   
    try:
        numero_fac = int(input("Introduce un número entero positivo"))
    except ValueError:
        numero_fac = int(input("Introduce un número entero positivo"))
    except RecursionError:
        numero_fac = int(input("Introduce un número entero positivo"))
    else:
        print(fac_recur(numero_fac))
        break
```

#### 7. Serie de Fibonacci

```python
num=10
lim=20

def fibonacci_iterativo(n):
    resultado = 0
    temp = 1
    for j in range (0,n+1) :
        print(f'n({j})={resultado}')
        temp,resultado = resultado,resultado + temp
        if resultado > lim:
            print(f'Se ha llegado al limite {lim}')
            break

fibonacci_iterativo(num)
```

#### 8. Verificación de Número Primo

```python
num=9

def es_primo(n):
    cont = 0
    for i in range (1,n+1):
        if(num%i==0):
            cont=cont+1
    return cont
result =es_primo(num)

if(result >2):
    print("No es un número primo")
else:
    print("Si es un número primo")
```

####  11. Programación Orientada a Objetos (POO) - Modelado de Especies


```python
class Animal:
    def __init__(self,nombre,tamaño):
        self.nombre = nombre
        self.tamaño = tamaño

    def ataque(self):
        print("Cada animal tiene un ataque diferente")

class Canino (Animal):
    def ataque(self):
        print("mordisco")
    def info(self):
        return "El canino {} tiene un tamaño de {}".format(self.nombre, self.tamaño)

class Felino (Animal):
    def ataque(self):
        print("rasgar")
    def trepar(self):
        print("tambien puedo trepar")

class Reptil (Animal):
    def ataque(self):
        print("veneno")
    def camuflar(self):
        print("ahora me ves... ahora ya no me ves...")

Perla = Canino("Perla",1.3)
Perla.info()
```

```python
Animal1 = Animal("pepe",1.8)
Animal1.ataque()
```

```python
Miau = Felino("Shaw",0.2)
Miau.trepar()
```

```python
Lizzard = Reptil("Sin nombre",2.3)
Lizzard.camuflar()
```
