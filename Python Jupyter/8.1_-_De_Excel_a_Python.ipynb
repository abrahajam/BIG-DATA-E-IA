## Merging o Vlookup

Vamos a juntar los 2 dataframes usando un left join

```python
rest_precio_df = rest_df.merge(precios_df,on=["Producto"], how = "left")
```

```python
rest_precio_df
```

Cuales son las diferencias entre merge de Python y Vlookup de Excel?

| Característica             | **Python (pandas.merge)**             | **Excel (BUSCARV / VLOOKUP)**           |
| -------------------------- | ------------------------------------- | --------------------------------------- |
| **Qué hace**               | Une tablas                            | Busca un valor en una tabla             |
| **Sentido de la búsqueda** | Izquierda, derecha o ambos            | Solo busca hacia la **derecha**         |
| **Varias claves**          | ✅ Sí (puede usar múltiples columnas)  | ❌ No (solo una clave)                   |
| **Velocidad**              | ⚡ Muy rápida incluso con muchos datos | 🐢 Lenta con grandes volúmenes de datos |
| **Automatización**         | ✅ Sí, con código reutilizable         | ❌ No, depende de fórmulas manuales      |
| **Resultado**              | Nueva tabla combinada                 | Un valor por celda                      |

## Tablas Dinámicas

Pandas también nos permite analizar datos mediante tablas dinámicas

Vamos creando una tabla dinámica que nos muestre las ventas por categoría

```python
#Juntar por categoria y sumar su valor
venxcat = pd.pivot_table(rest_precio_df, values = "Precio", index = "Categoria", aggfunc = np.sum) 
#(aggfunc = "sum") #rest_precios_df: La tabla base 
# values = "Precio": columna a resumir
# index = "Categoria" : columna para agrupar
#aggfunc = np.sum : sumar valores
```

```python
venxcat
```

Ahora una tabla dinámica que nos muestre las ventas por categoría y por tipo en la columna

```python
venxcat = pd.pivot_table(rest_precio_df, values = "Precio", index = "Categoria", columns="Tipo_x",aggfunc = np.sum) 
```

```python
venxcat
```

Crea una tabla dináica que muestre las ventas por tipo y categoría (filas), y por mesero (columnas)

```python
venxtipocatxmesero = pd.pivot_table(rest_precio_df, values = "Precio", index = ["Tipo_x", "Categoria"], columns = "Mesero", aggfunc="sum" )
```
