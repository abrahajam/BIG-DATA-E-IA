Vamos a juntar los 2 dataframes usando un left join

Cuales son las diferencias entre merge de Python y Vlookup de Excel?

| Característica             | **Python (pandas.merge)**             | **Excel (BUSCARV / VLOOKUP)**           |
| -------------------------- | ------------------------------------- | --------------------------------------- |
| **Qué hace**               | Une tablas                            | Busca un valor en una tabla             |
| **Sentido de la búsqueda** | Izquierda, derecha o ambos            | Solo busca hacia la **derecha**         |
| **Varias claves**          | ✅ Sí (puede usar múltiples columnas)  | ❌ No (solo una clave)                   |
| **Velocidad**              | ⚡ Muy rápida incluso con muchos datos | 🐢 Lenta con grandes volúmenes de datos |
| **Automatización**         | ✅ Sí, con código reutilizable         | ❌ No, depende de fórmulas manuales      |
| **Resultado**              | Nueva tabla combinada                 | Un valor por celda                      |
