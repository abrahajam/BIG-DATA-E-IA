```python
from selenium import webdriver #Controla el navegador(Chrome en este caso)
from selenium.webdriver.chrome.service import Service #Gestiona el servicio del driver en Chrome
from urllib.parse import urljoin # Sirve para usar una URL dinámica (Ej:Pasar de página)
from selenium.webdriver.chrome.options import Options #Permite configurar opciones de navegador (modo headless, etc..)
from webdriver_manager.chrome import ChromeDriverManager #Decarga y gestiona automaticamente la versión correcta del Driver
from selenium.webdriver.common.by import By #Sirve para localizar elementos en la página (por CSS,ID, etc)
from selenium.webdriver.support.ui import WebDriverWait #Permite hacer esperas explicitas hasta que algo esté presente / completado
from selenium.webdriver.support import expected_conditions as EC #Define las condiciones que esperamos (elementos, visibilidad, etc)

options = Options() #Configuración ordenador modo headless

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options = options)

titulos,stocks,precios =[],[],[]

url = "https://books.toscrape.com/"

try:
    while True:
        driver.get(url) #Cargar la página en el navegador
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "article.product_pod"))) #Espera explicita a que todos los valores "article.product_pod" estén cargados
        soup = BeautifulSoup(driver.page_source,"html.parser") # Guardar y procesar el HTML con selenium y beautifulsoup direcctamente
        articles = soup.select("article.product_pod") # Separar los componentes "article.product_pod"

        for libro in contenido_soup.find_all("article", attrs={"class":"product_pod"}):
            titulo = libro.h3.a["title"].strip()
            titulos.append(titulo)

            precio = libro.find("p",class_="price_color").get_text(strip=True)
            precios.append(precio)

            stock=libro.find("p",class_="instock availability").get_text(strip="True")
            stocks.append(stock)

        #Buscar el enlace a la siguiente pagina (si existe)
        next_link =soup.select_one("li.next > a")
        if next_link:
            url = urljoin(url,next_link.get("href"))
        else:
            break

    libros_df = pd.DataFrame({"Titulo":titulos,"Precio":precios,"Stocks":stocks})    

    libros_df.to_csv("libros_prueba.csv", index = False, encoding = "utf-8")
finally:
    driver.quit()
            
```
