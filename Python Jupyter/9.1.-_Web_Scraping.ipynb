## Extracción de Datos

Guarda el código HTML de la página en un objeto Contenido

```python
contenido = driver.page_source
```

```python
contenido
```

Conviertelo a BeautifulSoup

```python
contenido_soup = BeautifulSoup(contenido)
```

```python
contenido_soup
```

```python
contenido_soup.find_all("article",attrs={"class":"product_pod"})
```

Arma un for en el que vayas agregando a nuestras listas los datos necesarios con a.find, usa . append para ir agregando los datos a las listas

```python
for libro in contenido_soup.find_all("article",attrs={"class":"product_pod"}):
    titulos = libro.find("h3") #Donde están los datos del libro en el HTML
    titulo_sucio.append(titulos.text) #Solo el texto porque sino me cogerá toda la caca de HTML

    precios = libro.find("p", attrs={"class":"price_color"})
    precios_sucio.append(precios.text)  

    stocks = libro.find("p", attrs={"class":"instock availability"})
    stocks_sucio.append(stocks.text)   
```
